# Plots

```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Set better plotting style
plt.style.use('ggplot')
```

```python
# Function to plot stuff
def plotter(vae_data, skip_data, freebits_data, dropout_data, dropout, 
            title, xlabel, ylabel):
    """
    Accepts a bunch of data and plot settings and outputs the corresponding plot
    """
    
    fig, ax = plt.subplots()

    # Get curves on axis
    ax.plot(x, vae_val_elbo, label = "Standard VAE")
    ax.plot(x, skip_val_elbo, label = "Skip-VAE")
    ax.plot(x, freebits_val_elbo, label = "Freebits VAE")
    ax.plot(x, dropout_val_elbo, label = "Dropout VAE ({}%)".format(dropout))
    ax.legend()
    
    # Textual
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    # Save the figure
    # Get grey background color for poster
    #fig.savefig(os.path.join("img", "val_elbo.png"), facecolor='#cccccc')#, bbox_inches='tight')

    # Get white facecolor
    fig.savefig(os.path.join("img", "val_elbo.png"))#, facecolor='#cccccc')#, bbox_inches='tight')
```
